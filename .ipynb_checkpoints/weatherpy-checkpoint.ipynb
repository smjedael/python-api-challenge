# WeatherPy

## Analysis
 - The "Latitude vs. Max Temperature" chart has a bell-shape curve, with the highest temperatures between the -30 and 30 degree latitudes. This plot provides the most convincing evidence that it is hotter as one approaches the equator.
 - The vast majority of cities near the equator have a humidity near or above 60%, providing additional evidence.
 - Cloudiness and wind speed do not appear to be good indicators as to whether it gets hotter as one approaches the equator. 

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import requests
import json
import seaborn as sns
import time
from citipy import citipy as cp
from config import apikey
from datetime import datetime
```

## Initialize Variables

```python
#Create latitude and Longitude ranges
lat_range = np.arange(-90,91,10)
lng_range = np.arange(-180,181,10)

#City sample size
num_cities = 500

#Create arrays to store values
cities = []
city = []
country = []
lat = []
lng = []
max_temp = []
humidity = []
cloudiness = []
wind_mph = []

#Create OpenWeather Settings
url = "http://api.openweathermap.org/data/2.5/weather?"
units = 'imperial'

query_url = f'{url}appid={apikey}&units={units}&q='
```

## Generate Cities List and Perform API Calls

```python
#Initiate data retrieval and log
print('Beginning Data Retrieval')
print('----------------------------------------')
while len(cities) < num_cities:
    
    for y in range(0,len(lat_range)-1):
        lat_count = 0

        for x in range(0,len(lng_range)-1):     
            if len(cities) == num_cities:
                break
                
            #Attempt to select equitable number of cities for each latitude range
            elif lat_count > num_cities/(len(lat_range)-1):   
                break
            else:
                cp_lat = np.random.randint(lat_range[y],lat_range[y+1],size=1)    #Random Latitude coordinate
                cp_lng = np.random.randint(lng_range[x],lng_range[x+1],size=1)    #Random Longitude coordinate
                cp_city = cp.nearest_city(cp_lat, cp_lng)
                
                city_name = f'{cp_city.city_name},{cp_city.country_code}'       #Formatted for easy use with Open Weather API
                
                #Query for weather data
                time.sleep(1.5)               #Slow processing of queries to remain under limit for free account
                
                weather_data = requests.get(query_url + city_name).json()
                
                #Only select cities with complete set of desired values
                try:
                    ci = weather_data['name']
                    co = weather_data['sys']['country']
                    lt = weather_data['coord']['lat']
                    lg = weather_data['coord']['lon']
                    t = weather_data['main']['temp_max']
                    h = weather_data['main']['humidity']
                    c = weather_data['clouds']['all']
                    w = weather_data['wind']['speed']
                    
                    #Store city values if if they have not been added previously
                    if city_name not in cities:
                        cities.append(city_name)
                        city.append(ci)
                        country.append(co)
                        lat.append(lt)
                        lng.append(lg)
                        max_temp.append(t)
                        humidity.append(h)
                        cloudiness.append(c)
                        wind_mph.append(w)
                        lat_count += 1
                        print(f'Processing Record {len(cities)} of {num_cities} | {ci}, {co}')
                        print(f'{url}appid=***masked***&units={units}&q={city_name}')
                except KeyError:
                    pass
```

```python
#Create dataframe to store values
cities_dict = {'City':city,
               'Country':country,
               'Latitude':lat,
               'Longitude':lng,
               'Max Temp (F)':max_temp,
               'Humidity (%)':humidity,
               'Cloudiness (%)':cloudiness,
               'Wind Speed (mph)':wind_mph  
              }
df_cities = pd.DataFrame(cities_dict)

#Sort columns
df_cities = df_cities[['City','Country','Latitude','Longitude','Max Temp (F)',
                       'Humidity (%)','Cloudiness (%)','Wind Speed (mph)']]
df_cities.to_csv('WeatherPy_data.csv')
df_cities.head()
```

## City Latitude vs. Max Temperature

```python
#Plot Charts
current_date = datetime.today().strftime('%m/%d/%Y')

plt.figure(figsize=(10,6))
plt.scatter(df_cities['Latitude'],
            df_cities['Max Temp (F)'],
            marker='o',
            s=40,
            edgecolor='black',
            linewidths=0.75,
            alpha=0.85,
           )

plt.xticks(np.arange(-90,91,30))
plt.xlim(-95,95)
plt.xlabel('City Latitude')
plt.ylabel('Max Temperature (F)')
plt.title(f'City Latitude vs. Max Temperature ({current_date})')
sns.set_style('darkgrid')
sns.set_context('poster')
plt.show()
plt.savefig('LatvsMaxTemp.png')    
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df_cities['Latitude'],
            df_cities['Humidity (%)'],
            marker='o',
            s=40,
            edgecolor='black',
            linewidths=0.75,
            alpha=0.85,
           )

plt.xticks(np.arange(-90,91,30))
plt.xlim(-95,95)
plt.xlabel('City Latitude')
plt.ylabel('Humidity (%)')
plt.title(f'City Latitude vs. Humidity ({current_date})')
sns.set_style('darkgrid')
sns.set_context('poster')
plt.show()
plt.savefig('LatvsHumid.png')
```
